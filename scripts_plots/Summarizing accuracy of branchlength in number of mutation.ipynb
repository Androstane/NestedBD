```python
import dendropy 
import numpy as np
import random
#tree= dendropy.Tree.get(path="test.MCC", schema="nexus")
#tree.write(path = 'tt.nw', schema='newick')
import dendropy 
import numpy as np
import random
from ete3 import Tree
def func(x, a):
    return a * x
from scipy.optimize import curve_fit
import pandas as pd
import ptitprince as pt 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
path ='simulated/20cell_halfevents/rep0/BDtree_scaled.nw'
with open(path, 'r') as tree_file:
        gt_tree = tree_file.read().replace('\n', ';')
        #print(gt_tree)
        gt_tree = Tree(gt_tree, format = 3)
leaf_list = []
for leaf in gt_tree:
    leaf_list.append(leaf.name)
#tree = dendropy.Tree.get(path= path + '/gt.nw', schema='newick') # or whatever relevant format if not newick
#pdm = tree.phylogenetic_distance_matrix()
#pdm.write_csv(path + '/gt_pdm.csv')
```

```python
def pwd(tree):
    l = []
    for i in range(0, len(leaf_list)):
        for j in range(i+1, len(leaf_list)):
            leaf1, leaf2 = leaf_list[i], leaf_list[j]
            if leaf1 == 'leaf1' or leaf2 == 'leaf1':
                continue
            l.append(tree.get_distance(leaf1, leaf2))
    lmin, lmax = min(l), max(l)
    for i, val in enumerate(l):
        l[i] = (val-lmin) / (lmax-lmin)
    return l
gt_pwd = pwd(gt_tree)
```

```python
#ginkgo 
path = 'simulated/20cell_halfevents/'
MP, NJ, NestedBD = [], [], []
for rep in ['rep0', 'rep1','rep2','rep3','rep4','rep5,','rep6','rep7','rep8','rep9']:
    #print(rep)
    gt_tree = Tree(path + rep + '/BDtree_scaled.nw', format = 1) 
    gt_pwd = pwd(gt_tree)
    with open(path + rep + '/pdm/NJ.tree', 'r') as tree_file:
        tree = tree_file.read().replace('\n', ';')
        #print(gt_tree)
        tree = Tree(tree, format = 1)
        mp_pwd = pwd(tree)
        m = curve_fit(func,gt_pwd, mp_pwd)
        s = [(x-y)**2 for x, y in zip(gt_pwd, mp_pwd)]
        #print('MP:', np.linalg.norm(np.array(s)))
        MP.append(np.linalg.norm(np.array(s))/len(s))
        #MP.extend(s)
    with open(path + rep + '/pdm/MP.tree', 'r') as tree_file:
        tree = tree_file.read().replace('\n', ';')
        #print(gt_tree)
        tree = Tree(tree, format = 1)
        mp_pwd = pwd(tree)
        #m = curve_fit(func,gt_pwd, mp_pwd)
        s = [(x-y)**2 for x, y in zip(gt_pwd, mp_pwd)]
        NJ.append(np.linalg.norm(np.array(s))/len(s))
        #NJ.extend(s)
    with open(path + rep + '/MCCtree_scaled.nw', 'r') as tree_file:
        tree = tree_file.read().replace('\n', ';')
        #print(gt_tree)
        tree = Tree(tree, format = 1)
        mp_pwd = pwd(tree)
        #m = curve_fit(func,gt_pwd, mp_pwd)
        s = [(x - y)**2 for x, y in zip(gt_pwd, mp_pwd)]
        NestedBD.append(np.linalg.norm(np.array(s))/len(s))
        #NestedBD.extend(s)
d = []
for rep in MP:
    d.append(['MP', rep])
for rep in NJ:
    d.append(['NJ', rep])
for rep in NestedBD:
    d.append(['NestedBD', rep])
data_half = pd.DataFrame(d)
data_half.columns = ['Method', 'Normalized Euclidaen Distance']
```

```python
#ginkgo 
path = 'simulated/20cell/'
MP, NJ, NestedBD = [], [], []
for rep in ['rep0', 'rep1','rep2','rep3','rep4','rep5,','rep6','rep7','rep8','rep9']:
    #print(rep)
    gt_tree = Tree(path + rep + '/BDtree_scaled.nw', format = 1) 
    gt_pwd = pwd(gt_tree)
    with open(path + rep + '/pdm/NJ.tree', 'r') as tree_file:
        tree = tree_file.read().replace('\n', ';')
        #print(gt_tree)
        tree = Tree(tree, format = 1)
        mp_pwd = pwd(tree)
        m = curve_fit(func,gt_pwd, mp_pwd)
        s = [(x-y)**2 for x, y in zip(gt_pwd, mp_pwd)]
        #print('MP:', np.linalg.norm(np.array(s)))
        MP.append(np.linalg.norm(np.array(s))/len(s))
        #MP.extend(s)
    with open(path + rep + '/pdm/MP.tree', 'r') as tree_file:
        tree = tree_file.read().replace('\n', ';')
        #print(gt_tree)
        tree = Tree(tree, format = 1)
        mp_pwd = pwd(tree)
        #m = curve_fit(func,gt_pwd, mp_pwd)
        s = [(x-y)**2 for x, y in zip(gt_pwd, mp_pwd)]
        NJ.append(np.linalg.norm(np.array(s))/len(s))
        #NJ.extend(s)
    with open(path + rep + '/MCCtree_scaled.nw', 'r') as tree_file:
        tree = tree_file.read().replace('\n', ';')
        #print(gt_tree)
        tree = Tree(tree, format = 1)
        mp_pwd = pwd(tree)
        #m = curve_fit(func,gt_pwd, mp_pwd)
        s = [(x - y)**2 for x, y in zip(gt_pwd, mp_pwd)]
        NestedBD.append((np.linalg.norm(np.array(s))/len(s))*0.275)
        #NestedBD.extend(s)
d = []
for rep in MP:
    d.append(['MP', rep])
for rep in NJ:
    d.append(['NJ', rep])
for rep in NestedBD:
    d.append(['NestedBD', rep])
data = pd.DataFrame(d)
data.columns = ['Method', 'Normalized Euclidaen Distance']
import ptitprince as pt 
import seaborn as sns 
fig = plt.figure(figsize =(10,6))
sns.set_style("whitegrid")
sns.set_context("talk", font_scale=1.0, rc={"lines.linewidth": 2.0})
pt.RainCloud(data=data, x='Method', y='Normalized Euclidaen Distance', orient = 'h')
plt.ylabel('Method', fontsize = 18)
plt.xlabel('Euclidean distance between normalized true and inferred pairwise distance(each point represent a single dataset)', fontsize = 18)
#plt.xlim([0, 0.2])
```
